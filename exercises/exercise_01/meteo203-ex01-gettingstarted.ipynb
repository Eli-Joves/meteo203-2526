Simple sin plot using numpy and matplotlib. <br>
This is to test the environments

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate data
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Plot the data
plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```
